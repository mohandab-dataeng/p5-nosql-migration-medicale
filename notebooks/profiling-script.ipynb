#### 1 - Chargement des libriairies pour les traitement de DB xls

```python
import pandas as pd
```

#### 2 - Chargement des fichiers xls pour traitement

```python
# Charge TOUTES les feuilles -> retourne un dictionnaire {nom_feuille: DataFrame}
df_profile = pd.read_csv("../data/healthcare_dataset.csv")

# Voir les noms des feuilles disponibles
print(list(df_profile.keys()))
```

#### 3 - Extraire toutes les infos pour chaque colonnes 

```python
def renseigne(df,name):
    df_types = pd.DataFrame({
        'colonne': df.columns,
        'type': df.dtypes.values,
        'longueur_max': df.apply(lambda col: col.astype(str).str.len().max()).values,
        'non_null': df.notnull().sum().values,
        'nan': df.isnull().sum().values, 
        'valeur_unique':df.nunique().values,
    })
    return df_types

# Boucle sur chaque feuille
file = {"healthcare_dataset": df_profile}
for nom_feuille, df in file.items():
    display(f"\n--- Feuille : {nom_feuille} ---")
    display(renseigne(df,"healthcare_dataset"))
```
